```python
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import prepare_model_for_kbit_training, LoraConfig, get_peft_model, TaskType
import torch

model_id = "microsoft/phi-2"

tokenizer = AutoTokenizer.from_pretrained(model_id)
tokenizer.pad_token = tokenizer.eos_token

base_model = AutoModelForCausalLM.from_pretrained(
    model_id,
    load_in_4bit=True,
    torch_dtype=torch.float16,
    device_map="auto",
    trust_remote_code=True
)

base_model = prepare_model_for_kbit_training(base_model)

lora_config = LoraConfig(
    r=8,
    lora_alpha=32,
    target_modules=["q_proj", "v_proj", "k_proj", "o_proj"],
    lora_dropout=0.1,
    bias="none",
    task_type=TaskType.CAUSAL_LM
)

model = get_peft_model(base_model, lora_config)
model.print_trainable_parameters()
```

```python
#Loaded dataset and formatted for causal language modeling with classification prompt
from datasets import load_dataset

dataset = load_dataset("csv", data_files={
    "train": "/content/gdrive/MyDrive/data/train.csv",
    "validation": "/content/gdrive/MyDrive/data/validation.csv"
})

def format_for_clm(example):
    instruction = (
        "Classify the classification label group of the text into ONE of the following labels:\n"
        "'Targeted', 'Stereotype', 'Offensive/Provocative', 'Normal', 'not a joke'.\n"
        "Return ONLY the label.\n\n"
    )
    text = f"{instruction}Joke: {example['joke']}\n\nlabel: {example['label']}"
    return {
        "text": text,
        "label": example["label"]  # needed for evaluation
    }

formatted = dataset.map(format_for_clm)

```

```python
formatted["validation"][100]
#loads example
```

```python
# Define a tokenization function for each example in the dataset
def tokenize(example):
    # Tokenize the 'text' field with padding and truncation
    encodings = tokenizer(
        example["text"],               # Input text to tokenize
        padding="max_length",          # Pad sequences to the max_length
        truncation=True,               # Truncate sequences longer than max_length
        max_length=512                 # Set maximum sequence length
    )

    # Labels = input_ids (for causal language modeling)
    encodings["labels"] = encodings["input_ids"].copy()

    return encodings

# Apply the tokenization function to the entire dataset in batches
tokenized = formatted.map(tokenize, batched=True)
```

```python
#Set tokenizer pad token to EOS token to prevent padding-related issues during training


tokenizer.pad_token = tokenizer.eos_token
```

```python
#Initialized data collator for causal language modeling with MLM disabled

from transformers import DataCollatorForLanguageModeling

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False
)
```

```python
# Configured Trainer with training arguments and initialized training pipeline for Phi-2 with LoRA

from transformers import TrainingArguments, Trainer

# Define the training configuration for the model
training_args = TrainingArguments(
    output_dir="./phi2-lora-label2",           # Directory to save checkpoints and model outputs
    per_device_train_batch_size=4,              # Batch size per device for training
    per_device_eval_batch_size=4,               # Batch size per device for evaluation
    num_train_epochs=3,                         # Total number of training epochs
    logging_steps=20,                           # How often to log during training
    save_strategy="epoch",                      # Save model checkpoint at the end of every epoch
    eval_strategy="epoch",                      # Evaluate model performance at the end of every epoch
    fp16=True,                                  # Use 16-bit floating point precision for faster training
    save_total_limit=1,                         # Only keep 1 checkpoint to save disk space
    report_to=None,                             # Disable reporting to external tools (like wandb)
    push_to_hub=True,                           # Push the trained model to the Hugging Face Hub
    hub_model_id="3elix/phi2_humor_bias2",      # Name of the model repo on the Hub
)

# Create a Trainer instance that handles the training and evaluation loop
trainer = Trainer(
    model=model,                                # The LoRA-adapted Phi-2 model
    args=training_args,                         # Training arguments defined above
    train_dataset=tokenized_clean["train"],     # Training data (already tokenized and cleaned)
    eval_dataset=tokenized_clean["validation"], # Validation data for evaluation
    tokenizer=tokenizer,                        # Tokenizer used to process text inputs
    data_collator=data_collator                 # Data collator handles padding and batching
)

# Start training the model
trainer.train()

```

```python
from sklearn.metrics import classification_report
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel
from tqdm import tqdm
import torch
import re

# Valid label set
valid_labels = [
    'targeted', 'stereotype', 'offensive/provocative', 'normal', 'not a joke'
]

# Cleaning and mapping functions
def clean_prediction(pred):
    pred = pred.lower().strip()
    pred = re.sub(r'\s+', ' ', pred)
    parts = [p.strip() for p in pred.split("/") if p.strip()]
    deduped = list(dict.fromkeys(parts))
    return "/".join(deduped)

def map_to_valid_label(pred):
    pred = pred.lower()
    for label in valid_labels:
        if label in pred:
            return label
    return "other"

# Load tokenizer and base + LoRA adapter model
base_model = AutoModelForCausalLM.from_pretrained("microsoft/phi-2")
tokenizer = AutoTokenizer.from_pretrained("3elix/phi2_humor_bias2")
model = PeftModel.from_pretrained(base_model, "3elix/phi2_humor_bias2")
model = model.to("cuda" if torch.cuda.is_available() else "cpu")

# Run inference and collect predictions
true_labels = []
predicted_labels = []

# Iterate through each example in the validation set
for example in tqdm(formatted["validation"]):
    # Get the true label and joke text from the example
    true_target = example.get("label")
    joke_text = example.get("joke")

    # Skip this example if either the label or joke text is missing
    if true_target is None or joke_text is None:
        continue

    # Define the classification instruction for the model
    instruction = (
        "Classify the classification label group of the text into ONE of the following labels:\n"
        "'Targeted', 'Stereotype', 'Offensive/Provocative', 'Normal', 'not a joke'.\n"
        "Return ONLY the label.\n\n"
    )

    # Create the final prompt by combining instruction and joke
    prompt = f"{instruction}Joke: {joke_text}\n\nlabel:"

    # Tokenize the prompt and send it to the model's device
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    # Generate the model's response using the input prompt
    output = model.generate(
        **inputs,
        max_new_tokens=5,  # Limit output to 5 tokens to reduce hallucination
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.pad_token_id
    )

    # Get the position where the model starts generating new tokens
    input_len = inputs["input_ids"].shape[1]

    # Extract the generated portion of the output
    generated_ids = output[0][input_len:]

    # Decode the generated tokens into a string prediction
    prediction = tokenizer.decode(generated_ids, skip_special_tokens=True).strip()


    prediction = clean_prediction(prediction)
    prediction = map_to_valid_label(prediction)

    predicted_labels.append(prediction)
    true_labels.append(true_target.strip().lower())

# Final report
print(classification_report(true_labels, predicted_labels, labels=valid_labels))
```
